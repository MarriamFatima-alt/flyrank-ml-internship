## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# What makes the pattern too messy for an if-statement?
reason = (
    "Relevance depends on many interacting signals (position, past CTR, "
    "query intent, content match) whose relationship isn't linear or fixed. "
    "A hand-written rule (e.g. 'if position <= 3 then relevant') would ignore "
    "these interactions and break as soon as search behavior shifts, "
    "whereas an ML model can learn and adapt from observed data."
)

print(reason)
```
